```python
    # Convert Date to a date/time type.
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)

# Verify
print(data['Date'].head())
print("\nData types:")
print(data.dtypes)
```

```python
#convert quantity to integer but first change nulls for zeros
# Option 1: Fill nulls with 0, then convert
data['Quantity'] = data['Quantity'].fillna(0).astype(int)

# Check result
print(data['Quantity'].head())
print("\nData type:", data['Quantity'].dtype)
```

```python
# Fill nulls, then convert
data['TotalPrice'] = data['TotalPrice'].fillna(0.0).astype(float)
data.dtypes
```

```python
# Filter for Gadgets AND TotalPrice > 150
filtered = data[(data['Category'] == 'Gadgets') & (data['TotalPrice'] > 150)]

print(filtered)
```

```python
# Sort by TotalPrice descending
sorted_data = data.sort_values('TotalPrice', ascending=False)

print(sorted_data[['OrderID', 'Product', 'TotalPrice']].head(10))
```

```python
top_10 = data.sort_values('TotalPrice', ascending=False).head(10)
print(top_10[['OrderID', 'CustomerName', 'Product', 'TotalPrice']])
```

```python
last_5 = data.sort_values('Date', ascending=False).head(5)
print(last_5[['Date','OrderID', 'CustomerName', 'Product', 'TotalPrice']])
```

```python
# Calculate total sales per region
sales_by_region = data.groupby('Region')['TotalPrice'].sum()

print(sales_by_region)
```

```python
# Average TotalPrice per Product
avg_price = data.groupby('Product')['TotalPrice'].mean()

print(avg_price)
```

```python
# Count orders per customer
orders_per_customer = data['CustomerName'].value_counts()

print(orders_per_customer)
```
